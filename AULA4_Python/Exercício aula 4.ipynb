```python
import cv2
```

```python
from matplotlib import pyplot as plt

def print_img(img):
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

## 1) Escolha uma imagem de seu interesse e realize as operações solicitadas abaixo:

## a) Leitura da imagem

```python
img = cv2.imread( 'lena.png') 
```

## b) Mostre as dimensões (shape) e número de canais

```python
print(img.shape)
print(img.size)
```

## c) Insira um pequeno retângulo na posição 0,0, de dimensão (20,20)

```python
img[0:20, 0:20] = (0,0,255)
print_img(img)
```

## d) Faça o quarto quadrante da imagem ficar pintada de preto.

```python
pos = int(255/2)
img[pos:255, pos:255] = (0,0,0)
print_img(img)
```

## e) Rotacione a imagem em 75 graus

```python
(alt, lar) = img.shape[:2]
centro = (lar // 2, alt // 2) 
M = cv2.getRotationMatrix2D(centro, 75, 1.0) 
img_rotacionada = cv2.warpAffine(img, M, (lar, alt))
print_img(img_rotacionada)
```

## f) Faça um flip horizontal na imagem

```python
flip_horizontal = cv2.flip(img, 1)
print_img(flip_horizontal)
```

## g) Redimensione a imagem para largura de 200 pixels

```python
largura = img.shape[1]
altura = img.shape[0]
proporcao = float(altura/largura)
largura_nova = 200
altura_nova = int(largura_nova*proporcao)
tamanho_novo = (largura_nova, altura_nova)
img_redimensionada = cv2.resize(img, tamanho_novo, interpolation = cv2.INTER_AREA)
print_img(img_redimensionada)
```

## h) Salve a imagem no disco com o nome "nova_imagem.png"

```python
cv2.imwrite("nova_imagem.png", img)
```
