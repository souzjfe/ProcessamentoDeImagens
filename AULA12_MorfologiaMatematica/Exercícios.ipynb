### 1) Utilize o elemento estruturante abaixo:




```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))

def plot(imgs, titles, img_per_line = 3):
  plt.figure(figsize=(20, 20))
  for i, (img, title) in enumerate(zip(imgs, titles)):
    plt.subplot(img_per_line, img_per_line, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(title)

    
```

a) Aplique eros√£o e dilata√ß√£o na imagem img1.png dispon√≠vel no diret√≥rio. 

```python
img1 = cv2.imread('img1.png', cv2.IMREAD_GRAYSCALE)

erosion = cv2.erode(img1, kernel, iterations=1)
dilation = cv2.dilate(img1, kernel, iterations=1)

plot([img1, erosion, dilation], ["Original", "Eros√£o", "Dilata√ß√£o"])
```

b) Aplique eros√£o, dilata√ß√£o, fechamento na imagem img2.png

```python
img2 = cv2.imread('img2.png', cv2.IMREAD_GRAYSCALE)


erosion = cv2.erode(img2, kernel, iterations=1)
dilation = cv2.dilate(img2, kernel, iterations=1)
closing = cv2.morphologyEx(img2, cv2.MORPH_CLOSE, kernel)

plot([img2, erosion, dilation, closing],  ["Original", "Eros√£o", "Dilata√ß√£o", "Fechamento"], 4)
```

c) Aplique eros√£o e dilata√ß√£o na imagem img3.png

```python
img3 = cv2.imread('img3.png', cv2.IMREAD_GRAYSCALE)


erosion = cv2.erode(img3, kernel, iterations=1)
dilation = cv2.dilate(img3, kernel, iterations=1)

plot([img3, erosion, dilation],  ["Original", "Eros√£o", "Dilata√ß√£o"])
```

d) Aplique eros√£o e dilata√ß√£o usando outro elemento estruturante nestas mesmas imagens. (Defina o kernel usado)

```python
img1 = cv2.imread('img1.png', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('img2.png', cv2.IMREAD_GRAYSCALE)
img3 = cv2.imread('img3.png', cv2.IMREAD_GRAYSCALE)

kernelEl = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
print(kernelEl)
erosion1 = cv2.erode(img1, kernelEl, iterations=1)
dilation1 = cv2.dilate(img1, kernelEl, iterations=1)


erosion2 = cv2.erode(img2, kernelEl, iterations=1)
dilation2 = cv2.dilate(img2, kernelEl, iterations=1)


erosion3 = cv2.erode(img3, kernelEl, iterations=1)
dilation3 = cv2.dilate(img3, kernelEl, iterations=1)
titles = np.tile(["Original", "Eros√£o", "Dilata√ß√£o"],3)
plot([img1, erosion1, dilation1,img2, erosion2, dilation2,img3, erosion3, dilation3], titles)
```

### 2) Considere a imagem img4.png

a) Aplique o filtro morfol√≥gigo de eros√£o (3x3). Qual √© o efeito na imagem?


```python
img4 = cv2.imread('img4.png', cv2.IMREAD_GRAYSCALE)


erosion = cv2.erode(img4, kernel, iterations=1)

plot([img4, erosion],  ["Original", "Eros√£o"], 2)
```

>Na imagem acima pode-se observar maior distancimento dos pixels ap√≥s a eros√£o.

b) Aplique o filtro morfol√≥gigo de dilata√ß√£o (3x3). Qual √© o efeito na imagem?

```python
img4 = cv2.imread('img4.png', cv2.IMREAD_GRAYSCALE)


dilation = cv2.dilate(img4, kernel, iterations=1)

plot([img4, dilation],  ["Original", "Dilata√ß√£o"], 2)
```

>Na imagem acima pode-se observar o agrupamento dos pixels ap√≥s a eros√£o.

c) Aplique o filtro morfol√≥gigo de abertura (3x3). Qual √© o efeito na imagem?

```python
img4 = cv2.imread('img4.png', cv2.IMREAD_GRAYSCALE)


opening = cv2.morphologyEx(img4, cv2.MORPH_OPEN, kernel)

plot([img4, opening],  ["Original", "Abertura"], 2)
```

>Na imagem acima menor quantidade de pixels brancos de pequenos tamanhos.

d) Aplique o filtro morfol√≥gigo de fechamento (3x3). Qual √© o efeito na imagem?

```python
img4 = cv2.imread('img4.png', cv2.IMREAD_GRAYSCALE)


closing = cv2.morphologyEx(img4, cv2.MORPH_CLOSE, kernel)

plot([img4, closing],  ["Original", "Fechamento"], 2)
```

>Na imagem acima pode se observar o preenchimento dos pixels "vazios" proximos aos pixels brancos

### 3) Como podemos obter as bordas dos objetos na imagem img5.png?

```python
img5 = cv2.imread('img5.png', cv2.IMREAD_GRAYSCALE)
threshold = cv2.adaptiveThreshold(img5, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
edges = np.absolute(255 - threshold)

edges = cv2.erode(edges, cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3)), iterations=1)
plot([img5, edges],  ["Original", "Bordas"], 2)
```
