```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage.filters import roberts, prewitt, sobel

img =  cv2.cvtColor(cv2.imread("imagemA.jpg"), cv2.COLOR_BGR2GRAY) 

def plot_comparar_com(imgagem_comparada, title = ''):
  plt.figure(figsize=(40, 20))
  plt.subplot(1, 4, 1)
  plt.imshow(img, cmap='gray')
  plt.title("Original")
  plt.xticks([])
  plt.yticks([])
  plt.subplot(1, 4, 2)
  plt.imshow(imgagem_comparada, cmap='gray')
  plt.title(title)
  plt.xticks([])
  plt.yticks([])  
  plt.show()
```

1) Implemente o filtro de suavização pela média da vizinhança

```python

key = 5
kernel = np.ones((key,key),np.float32)/key**2
smoothed = cv2.filter2D(img,-1,kernel)

plot_comparar_com(smoothed, "Vizinhança média")
```

8) Escolha duas imagens, realize a adição destas. 

```python
def add_images(image1, image2):
    assert image1.shape == image2.shape
    
    image1 = image1.astype(np.float32)
    image2 = image2.astype(np.float32)
    
    added_image = cv2.add(image1, image2)
    
    return added_image.astype(np.uint8)
```

```python
com =  cv2.imread("com.jpg")
sem =  cv2.imread("sem.jpg")


plt.figure(figsize=(40, 20))
plt.subplot(1, 4, 1)
plt.imshow(add_images(sem, com), cmap='gray')
plt.title("Original")
plt.xticks([])
plt.yticks([])
plt.show()
```

9) Escolha duas imagens, realize a subtração destas. O ideal é para representar a remoção de fundo.

```python
def subtract_images(image1, image2):
    assert image1.shape == image2.shape
    
    image1 = image1.astype(np.float32)
    image2 = image2.astype(np.float32)
    
    subtracted_image = cv2.subtract(image1, image2)
    
    return subtracted_image.astype(np.uint8)
```

```python
com =  cv2.imread("com.jpg")
sem =  cv2.imread("sem.jpg")


plt.figure(figsize=(40, 20))
plt.subplot(1, 4, 1)
plt.imshow(subtract_images(sem, com), cmap='gray')
plt.title("Original")
plt.xticks([])
plt.yticks([])
plt.show()
```
